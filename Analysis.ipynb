```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/Users/pavani_19/Business_Analyst_Project/marketing_campaign.csv',sep=';')
print(df.shape)
```

```python
edu_counts = df['Education'].value_counts()

plt.figure(figsize=(7, 4))
plt.bar(edu_counts.index, edu_counts.values, color='steelblue')
plt.title('Customer Education Level')
plt.xlabel('Education')
plt.ylabel('Number of Customers')
plt.tight_layout()
plt.show()
```

```python
marital_counts = df['Marital_Status'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(marital_counts.values, labels=marital_counts.index, autopct='%1.1f%%')
plt.title('Marital Status of Customers')
plt.show()
```

```python
campaign_cols = ['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5']

df[campaign_cols].sum().plot(kind='bar', color='mediumpurple')
plt.title('Campaign Acceptance Count')
plt.xlabel('Campaign')
plt.ylabel('Number of Customers Who Accepted')
plt.tight_layout()
plt.show()
```

```python
channel_cols = ['NumDealsPurchases','NumWebPurchases','NumCatalogPurchases','NumStorePurchases']

df[channel_cols].sum().plot(kind='bar', color='dodgerblue')
plt.title('Total Purchases by Channel')
plt.xlabel('Channel')
plt.ylabel('Total Number of Purchases')
plt.tight_layout()
plt.show()
```

```python
# Seaborn chart 1 — Birth Year Distribution
sns.histplot(df['Year_Birth'], kde=True)
plt.title("Customer Birth Year Distribution")
plt.show()
```

```python
# Seaborn chart 2 — Income vs Spending (shows relationship)
sns.scatterplot(x=df['Income'], y=df['MntWines'])
plt.title("Income vs Wine Spending")
plt.xlabel("Income")
plt.ylabel("Amount Spent on Wines")
plt.show()
```

```python
# NumPy — Basic stats summary (shows you know numpy)
print("Income Stats:")
print("Mean   :", np.mean(df['Income'].dropna()))
print("Median :", np.median(df['Income'].dropna()))
print("Std Dev:", np.std(df['Income'].dropna()))
```
